# DRISHTI Weekly Retraining Pipeline

This notebook mirrors the production retraining script in `backend/ml/weekly_retraining_pipeline.py`. It uses the Astram event dataset plus Supabase `retraining_prediction_dataset` rows generated from citizen complaints, predictions, and post-event feedback.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
BACKEND_ROOT = PROJECT_ROOT / 'backend'
sys.path.append(str(BACKEND_ROOT))

from ml.weekly_retraining_pipeline import load_astram_dataset, load_retraining_dataset, train_models, save_artifacts
```

```python
csv_path = PROJECT_ROOT / 'Astram event data_anonymized - Astram event data_anonymizedb40ac87 (2).csv'
historical_df = load_astram_dataset(csv_path)
historical_df.head()
```

```python
retraining_df = load_retraining_dataset()
retraining_df.shape
```

```python
import pandas as pd
model_df = pd.concat([historical_df, retraining_df], ignore_index=True)
model_df[['duration_minutes', 'impact_level']].describe(include='all')
```

```python
result = train_models(model_df)
result['metrics']
```

```python
save_artifacts(result, BACKEND_ROOT / 'app' / 'models' / 'candidates')
```
